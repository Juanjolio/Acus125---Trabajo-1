## Caso 1: Subamortiguado

#### $0 < \zeta < 1$

Cuando $0 < \zeta < 1$, tratamos con un caso de vibración libre subamortiguada, por lo que las raíces $\alpha$ son de la forma:

\begin{align*}
\alpha \quad = \quad \omega_{n} \left(-\zeta ± j\sqrt{1-\zeta^2}\right)
\end{align*}

Para hacer más cómoda la lectura, definimos $D = \sqrt{1-\zeta^2}$, por lo que la ecuación anterior nos queda expresada como:

\begin{align*}
\alpha \quad = \quad \omega_{n} \left(-\zeta ± jD\right)
\end{align*}

Luego, al reemplazarlo en la solución general tenemos que:

\begin{align*}
X(t) \quad &= \quad B_1 e^{\left(-\zeta - jD\right)} + B_1 e^{\left(-\zeta + jD\right)}\\
&= \quad e^{-\zeta\omega_n t} \left[B_1 cos\left(\omega_n D t\right) - j B_1 sen\left(\omega_n D t\right) + B_2 cos\left(\omega_n D t\right) + B_2 sen\left(\omega_n D t\right)\right]\\
&= \quad e^{-\zeta\omega_n t}\left[(B_1+B_2) cos\left(\omega_n D t\right) + j (B_2 - B_1) sen\left(\omega_n D t\right)\right]\\
\end{align*}

Como $B_1 y B_2$ son constantes, podemos decir que $B_1 + B_2 = C_1$ y que $j(B_2 - B_1) = C_2 $, por lo que la ecuación de solución general del sistema nos queda como sigue:

\begin{align*}
X(t) \quad &= \quad e^{-\zeta\omega_n t}\left[C_1 cos\left(\omega_n D t\right) + C_2 sen\left(\omega_n D t\right)\right]\\\tag{1}
\end{align*}

A continuación, definimos las condiciones iniciales del sistema, para así poder obtener las constantes mencionadas anteriormente:

\begin{align*}
X(0) \quad = \quad X_0 \qquad\qquad\qquad \dot{X}(0) \quad = \quad \dot{X}_0
\end{align*}

Primero, analizamos con la condición inicial cuando t=0 :

\begin{align*}
X(0) \quad &= \quad e^{-\zeta\omega_n*0}\left[C_1 cos\left(\omega_n D *0\right) + C_2 sen\left(\omega_n D *0\right)\right]\\
X(0) \quad &= \quad e^{0}\left[C_1 cos\left(0\right) + C_2 sen\left(0\right)\right]\\
X(0) \quad &= \quad \left[C_1  + C_{2} * 0\right]\\
X(0) \quad &= \quad C_1\\
\end{align*}

Luego, al reemplazarlo el valor de $C_1$ en la ecuación (1) tenemos que:

\begin{align*}
X(t) \quad &= \quad e^{-\zeta\omega_n t}\left[X_{0} cos\left(\omega_n D t\right) + C_2 sen\left(\omega_n D t\right)\right]\\\tag{2}
\end{align*}

Entonces, derivamos la ecuación (2) para posteriormente intentar obtener el valor de la constante $C_{2}$:

\begin{align*}
\dot{X}(t) = &-\zeta\omega_n e^{-\xi\omega_n t}\left[X_{0} cos\left(\omega_n D t\right) + C_2 sen\left(\omega_n D t\right)\right] + e^{-\zeta\omega_n t}\left[-X_{0}\omega_n D sen\left(\omega_n D t\right) + C_{2}\omega_n D cos\left(\omega_n D t\right)\right]\\\tag{3}
\end{align*}

Luego, analizamos la ecuación (3) obtenida para cuando t=0:

\begin{align*}
\dot{X}(0) = &-\zeta\omega_n e^{-\zeta\omega_n*0}\left[X_{0} cos\left(\omega_n D *0\right) + C_2 sen\left(\omega_n D *0\right)\right]\\ &+ e^{-\zeta\omega_n*0}\left[-X_{0}\omega_n D sen\left(\omega_n D *0\right) + \omega_{n}C_{2}\omega_n D cos\left(\omega_n D *0\right)\right]\\
\dot{X}_0 = &-\zeta\omega_n e^{0}\left[X_{0} cos\left(0\right) + C_2 sen\left(0\right)\right] + e^{0}\left[-X_{0}\omega_n D sen\left(0\right) + \omega_{n}C_{2}\omega_n D cos\left(0\right)\right]\\
\dot{X}_0 = &-\zeta\omega_n X_{0} + \omega_{n}C_{2}\omega_n D\\
\end{align*}

Despejamos $C_{2}$:
\begin{align*}
&\Rightarrow \qquad e^{0}\omega_{n}C_{2}\omega_n D &= \quad &\dot{X}_0 + \zeta\omega_n X_{0}\\
&\Rightarrow \qquad C_{2} &= \quad &\frac{\dot{X}_0 + \zeta\omega_n X_{0}}{\omega_n D}\\
&\Rightarrow \qquad C_{2} &= \quad &\frac{\dot{X}_0 + \zeta\omega_n X_{0}}{\omega_n\sqrt{1-\zeta^2}}\\
\end{align*}

y reemplazamos $D$ y $C_2$ en la ecuación (2) nos queda la solución general de la siguiente forma:

\begin{align*}
X(t) \quad &= \quad e^{-\zeta\omega_n t}\left[X_{0} cos\left(\omega_n \sqrt{1-\xi^2} t\right) + \frac{\dot{X}_0 + \zeta\omega_n X_{0}}{\omega_n\sqrt{1-\xi^2}} sen\left(\omega_n \sqrt{1-\zeta^2} t\right)\right]\tag{4}\\
\end{align*}

Luego, suponemos que la solución de esta nueva ecuación (4) es de la forma:

\begin{align*}
X(t) \quad &= A e^{-\zeta\omega_n t} sen(\omega_d t + \phi_d)\\
&= A e^{-\zeta\omega_n t} \left[ sen(\omega_d t) cos(\phi_d) + cos(\omega_d t) sen(\phi_d)  \right]\\
&= A e^{-\zeta\omega_n t}sen(\omega_d t) cos(\phi_d) + A e^{-\zeta\omega_n t} cos(\omega_d t) sen(\phi_d)\tag{5}\\
\end{align*}

donde $\omega_d = \omega_n \sqrt{1-\zeta^2}$. Por lo que analizamos con el valor de la condición inicial cuando t=0 :

\begin{align*}
X(0) \quad &= A e^{-\zeta\omega_n*0}sen(\omega_d*0) cos(\phi_d) + A e^{-\zeta\omega_n*0} cos(\omega_d*0) sen(\phi_d)\\
&= A e^{0}sen(0)cos(\phi_d) + A e^{0} cos(0) sen(\phi_d)\\
X_0 \quad &= A sen(\phi_d)\tag{6}\\ 
\end{align*}

Luego, al derivar la ecuación (5) tenemos que:

\begin{align*}
\dot{X}(t) \quad &= -\zeta\omega_n A e^{-\zeta\omega_n t}sen(\omega_d t) cos(\phi_d) + A \omega_d e^{-\zeta\omega_n t}cos(\omega_d t) cos(\phi_d)+ -\zeta\omega_n A e^{-\zeta\omega_n t} cos(\omega_d t) sen(\phi_d) - A \omega_d e^{-\zeta\omega_n t} sen(\omega_d t) sen(\phi_d)\\
\dot{X}(t) \quad &= A e^{-\zeta\omega_n t}cos(\omega_d t)\left[\omega_d cos(\phi_d)- \zeta\omega_n sen(\phi_d)\right]
- A e^{-\zeta\omega_n t} sen(\omega_d t)\left[\zeta\omega_n cos(\phi_d) + \omega_d sen(\phi_d)\right]\tag{7}\\ 
\end{align*}

A continuación, analizamos esta ecuación (7) con la condición inicial t=0 :

\begin{align*}
\dot{X}(0) \quad &= A e^{-\zeta\omega_n *0}cos(\omega_d*0)\left[\omega_d cos(\phi_d)- \zeta\omega_n sen(\phi_d)\right]
- A e^{-\zeta\omega_n *0} sen(\omega_d *0)\left[\zeta\omega_n cos(\phi_d) + \omega_d sen(\phi_d)\right]\\
\dot{X}_0 \quad &= A e^{0}cos(0)\left[\omega_d cos(\phi_d)- \zeta\omega_n sen(\phi_d)\right]
- A e^{0} sen(0)\left[\zeta\omega_n cos(\phi_d) + \omega_d sen(\phi_d)\right]\\
\dot{X}_0 \quad &= A \left[\omega_d cos(\phi_d)- \zeta\omega_n sen(\phi_d)\right] \tag{8}\\
\end{align*}
 
 y de la ecuación (6) despejamos que $sen(\phi_d) = \frac{X_0}{A}$ , por lo que reemplazando tenemos:
 
 \begin{align*}
\dot{X}_0 \quad &= \quad A \left[\omega_d cos(\phi_d)- \zeta\omega_n \frac{X_0}{A}\right]\\
&= \quad A \omega_d cos(\phi_d)- \zeta\omega_n X_0\\
\Rightarrow \quad \dot{X}_0 + \zeta\omega_n X_0 \quad &=\quad A \omega_d cos(\phi_d)\\
\Rightarrow \quad\frac{\dot{X}_0 + \zeta\omega_n X_0}{\omega_d}\quad &= \quad A cos(\phi_d)\\
\frac{\dot{X}_0 + \zeta\omega_n X_0}{\omega_n \sqrt{1-\zeta^2}} \quad &= \quad A cos(\phi_d)\tag{9}\\
\end{align*}




Luego, para encontrar el valor de la constante A, analizamos la siguiente figura extraída de las ecuaciones (6) y (9):

<p style="text-align:center">
	<img src="Triángulo phid.JPEG" alt="Triángulo phid" width=65%/></a><br>
    <strong> Figura 1: Análisis gráfico de las ecuaciones (6) y (9).</strong>
</p>

Por lo que tenemos que para A :

\begin{align*}
A = \sqrt{ X_0 ^2 + \left( \frac{\dot{X}_0 t + \zeta\omega_n X_0}{\omega_n \sqrt{1-\zeta ^2}}\right)^2}
\end{align*}

Por otro lado, para $\phi_d$ :

\begin{align*}
&sen(\phi_d) = \frac{X_0}{A} \qquad \qquad &y \qquad \qquad cos(\phi_d) = \frac{ \frac{ \dot{X}_0 + \zeta\omega_n X_0}{\omega_n \sqrt{1-\zeta^2}}}{A}\\
\end{align*}

\begin{align*}
\Rightarrow tan(\phi_d) &= \frac{sen(\phi_d)}{cos(\phi_d)}\\
&= \frac{X_0}{ \frac {\dot{X}_0 + \zeta\omega_n X_0}{\omega_n \sqrt{1-\zeta^2}}}\\
&= \frac{X_0 \omega_n \sqrt{1-\zeta^2}}{\dot{X}_0 + \zeta\omega_n X_0}\\
\Rightarrow \qquad\phi_d &= tan^{-1} \left( \frac{X_0 \omega_n \sqrt{1-\zeta^2}}{\dot{X}_0 + \zeta\omega_n X_0} \right)\\
\end{align*}

Por lo tanto, tenemos que la ecuación (5) : $ X(t) = A e^{-\zeta\omega_n t} sen(\omega_d t + \phi_d) $, al reemplazar los valores $\omega_d$, $A$ y $\phi_d$ nos queda como sigue:

\begin{align*}
X(t) \quad &= \quad \sqrt{ X_0 ^2 + \left( \frac{\dot{X}_0 t + \zeta\omega_n X_0}{\omega_n \sqrt{1-\zeta ^2}}\right)^2} e^{-\zeta\omega_n t} sen\left(\omega_n \sqrt{1-\zeta^2} t + tan^{-1} \left( \frac{X_0 \omega_n \sqrt{1-\zeta^2}}{\dot{X}_0 + \zeta\omega_n X_0} \right)\right)\\
\end{align*}




```python
import numpy as np
import matplotlib.pyplot as plt

#CASO 1  0<zeta<1 Subamortiguado
#Creamos un arreglo con valores temporales
t=np.linspace(0,10,1000)
#Para fines ilustrativos asumimos algunos valores de rigidez, masa, y amortiguamiento del sistema 
k=500
m=100
z=0.1
w0=np.sqrt(k/m)
#También asumimos unos valores predeterminados de posición y velocidad inicial
x0=3
v0=9
#Obtenemos la frecuencia natural amortiguada
wd=w0*np.sqrt(1-z**2)
#Asi obtenemos la Amplitud
A=np.sqrt(x0*2+((v0+z*w0*x0)/wd)*2)
#Obtenemos phi
phi=np.arctan((x0*wd)/(v0+z*w0*x0))
#Obtenemos el desplazamiento
x1=A*np.exp(-z*w0*t)*np.sin(wd*t-phi)
#Obtenemos la exponencial que envuelve el comportamiento del sistema
env1=A*np.exp(-z*w0*t)
env2=-A*np.exp(-z*w0*t)
plt.plot(t,x1,color='k')
plt.plot(t,env1,color='r',linestyle='--')
plt.plot(t,env2,color='r',linestyle='--')
plt.xlabel('tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Sistema Subamortiguado');
```

## Caso 2: Críticamente amortiguado

#### $\zeta=1$

Cuando el valor de $\zeta$ es igual a 1, nos encontramos con el caso críticamente amortiguado, en el cual la ecuación característica tiene una única raíz real.  

\begin{align}
\alpha= \omega_{n} \left(-\zeta ± j\sqrt{1-\zeta^2}\right)
\end{align}  

Lo que está dentro de la raíz, es decir $1-\zeta^{2}$ es igual a 0, por lo que se elimina esa raíz cuadrada, y al mismo tiempo la parte imaginaria, quedando finalmente:  
\begin{align}
\alpha= -\zeta\omega_{n} = -\omega_{n}
\end{align}  

Reemplazando este valor en la solución general tenemos que la primera solución a la ecuación diferencial es:

\begin{align}
e^{-\omega_{n}t}
\end{align}  

Para obtener una segunda solución linealmente independiente se debe multiplicar por $t$ la primera, obteniendo así:  

\begin{align}
t e^{-\omega_{n}t}
\end{align}

Posteriormente se llega a que la solución general está dada por la siguiente expresión:

\begin{align}
X(t)=A_{1}e^{-\omega_{n}t}+A_{2}t e^{-\omega_{n}t}
\end{align}

Factorizando por $e^{-\omega_{n}t}$ se tiene:

\begin{align}
X(t)=e^{-\omega_{n}t}(A_{1}+A_{2}t)             
\end{align}

Ahora se aplican las condiciones iniciales:

\begin{align}
X(0)=X_{0} && \dot{X}(0)=\dot{X}_{0}           
\end{align}

Primero reemplazamos $t=0$ en $X(t)$:

\begin{align}
X(0)=e^{-\omega_{n}*0}(A_{1}+A_{2}*0)= A_1 = X_{0}        
\end{align}

Ahora se debe derivar $X(t)$, obteniendo $\dot{X}(t)$ y así reemplazar en $t=0$ para obtener el valor de $A_{2}$:

\begin{align}
\dot{X}(t)= -\omega_{n}e^{-\omega_{n}t}(A_{1}+A_{2}t)+A_{2}e^{-\omega_{n}t}       
\end{align}

Ahora reemplazando $t=0$ se tiene:

\begin{align}
\dot{X}(0)=-A_{1}\omega_{n}+A_{2}       
\end{align}

Pero sabemos que $A_{1}=X_{0}$, por lo que:  

\begin{align}
\dot{X}(0)= -X_{0}\omega_{n}+A_{2}=\dot{X}_{0}      
\end{align}

Y así obtenemos el valor de $A_{2}$:
\begin{align}
A_{2}=\dot{X}{0}+\omega{n}X_{0}    
\end{align}

Y de esta manera llegamos finalmente a la solución general:

\begin{align}
X(t)= e^{-\omega_{n}t}[X_{0}+(\dot{X}{0}+\omega{n}X_{0})t]
\end{align}

```python
#Caso 2 zeta=1 Críticamente amortiguado
t=np.linspace(0,10,1000)
#Definimos valores de rigidez y masa del sistema
k=500
m=100
#Definimos una posición y velocidad inicial
x0=3
v0=9
#Obtenemos la frecuencia natural del sistema
w0=np.sqrt(k/m)
#Y así obtenemos la fórmula para el desplazamiento
x2=np.exp(-w0*t)*(x0+((v0+w0*x0)*t))

plt.plot(t,x2,color='k')
plt.xlim((0, 10))
plt.xlabel('tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Sistema Críticamente amortiguado');
```

## Caso 2: Sobre amortiguado

#### $\zeta > 1$

Cuando el valor de $\zeta$ es mayor que 1, hablamos de un sistema sobre amortiguado, donde las raíces de la ecuación característica son:
    
\begin{align*}
\alpha = \omega_n \left(-\zeta ± \sqrt{\zeta ^2 - 1}  \right)
\end{align*}
 
Por lo que la solución general nos queda de la forma:
\begin{align*}
X(t) = C_1 e^{-\omega_n \left(\zeta + \sqrt{\zeta ^2 - 1}  \right)t} + C_2 e^{-\omega_n \left(\zeta - \sqrt{\zeta ^2 - 1}  \right)t}\tag{10}
\end{align*}

Luego, buscamos los valores de las constantes $C_1$ y $C_2$ a través de las condiciones iniciales :

\begin{align*}
X(0) = X_0 \qquad \qquad y \qquad \qquad \dot{X}(0) = \dot{X}_0
\end{align*}

Por lo que analizamos la ecuación (10) cuando $t=0$ :

\begin{align*}
X(0) &= C_1 e^{-\omega_n \left(-\zeta + \sqrt{\zeta ^2 - 1}  \right)*0} + C_2 e^{-\omega_n \left(-\zeta - \sqrt{\zeta ^2 - 1}  \right)*0}\\
X(0) &= C_1 e^0 + C_2 e^0\\
X_0 &= C_1 + C_2\tag{11}\\
\end{align*}

Luego, derivamos la ecuación (10), para poder analizar la otra condición inicial:

\begin{align*}
\dot{X}(t) &= -\omega_n \left(\zeta + \sqrt{\zeta ^2 - 1}  \right) C_1 e^{-\omega_n \left(\zeta + \sqrt{\zeta ^2 - 1}  \right)t} -\omega_n \left(\zeta - \sqrt{\zeta ^2 - 1}  \right) C_2 e^{-\omega_n \left(\zeta - \sqrt{\zeta ^2 - 1}  \right)t}
\end{align*}

Analizamos cuando $t=0$ y tenemos que :

\begin{align*}
\dot{X}(t) &= -\omega_n \left(\zeta + \sqrt{\zeta ^2 - 1}  \right) C_1 e^{-\omega_n \left(\zeta + \sqrt{\zeta ^2 - 1}  \right)*0} -\omega_n \left(\zeta - \sqrt{\zeta ^2 - 1}  \right) C_2 e^{-\omega_n \left(\zeta - \sqrt{\zeta ^2 - 1}  \right)*0}\\
\dot{X}(t) &= -\omega_n \left(\zeta + \sqrt{\zeta ^2 - 1}  \right) C_1 e^{0} -\omega_n \left(\zeta - \sqrt{\zeta ^2 - 1}  \right) C_2 e^{0}\\
\dot{X}(t) &= -\omega_n \left(\zeta + \sqrt{\zeta ^2 - 1}  \right) C_1 -\omega_n \left(\zeta - \sqrt{\zeta ^2 - 1}  \right) C_2\tag{12}\\
\end{align*}

Luego, tenemos dos ecuaciones, (11) y (12), y dos incógnitas, $C_1$ y $C_2$, por lo que podemos utilizar el método de Cramer para encontrar dichas constantes del siguiente sistema:

\begin{align*}
\left[\begin{matrix} 1 & 1\\ -\omega_n \left(\zeta + \sqrt{\zeta ^2 - 1}  \right) & -\omega_n \left(\zeta - \sqrt{\zeta ^2 - 1}  \right)\\ \end{matrix}\right] \left[\begin{matrix} C_1\\ C_2\\ \end{matrix}\right] = \left[\begin{matrix} X_0\\ \dot{X}_0\\ \end{matrix}\right]
\end{align*}

\begin{align*}
\Rightarrow \qquad C_{1} &= \frac{\left[\begin{matrix} X_0 & 1\\ \dot{X}_0 & -\omega_n \left(\zeta - \sqrt{\zeta ^2 - 1}  \right)\\ \end{matrix}\right]}{\left[\begin{matrix} 1 & 1\\ -\omega_n \left(\zeta + \sqrt{\zeta ^2 - 1}  \right) & -\omega_n \left(\zeta - \sqrt{\zeta ^2 - 1}  \right)\\ \end{matrix}\right] }
= \frac{-X_0 \omega_n \left(\zeta - \sqrt{\zeta ^2 - 1}  \right)- \dot{X}_0} {2 \omega_n \sqrt{\zeta^2 - 1} }
\end{align*}

\begin{align*}
\Rightarrow \qquad C_{2} &=\frac{\left| \begin{matrix} 1 & X_0\\ -\omega_n \left( \zeta + \sqrt{\zeta^2 - 1} \right) & \dot{X}_0 \\  \end{matrix}\right|}{\left| \begin{matrix}   1 & 1\\  -\omega_n \left(\zeta + \sqrt{\zeta^2 - 1}\right)  & -\omega_n \left(\zeta - \sqrt{\zeta ^2 - 1}\right)\\   \end{matrix}\right|} = \frac{ \dot{X}_0 + X_0 \omega_n \left(\zeta + \sqrt{\zeta^2 - 1}  \right) }{2 \omega_n \sqrt{\zeta^2 - 1}}
\end{align*}

Por lo tanto, reemplazando $C_1$ y $C_2$, la ecuación general (10) nos queda :

\begin{align*}
X(t) &= \frac{-X_0 \omega_n \left(\zeta - \sqrt{\zeta ^2 - 1}  \right)- \dot{X}_0} {2 \omega_n \sqrt{\zeta^2 - 1} } e^{-\omega_n \left(\zeta + \sqrt{\zeta ^2 - 1}  \right)t} + \frac{ \dot{X}_0 + X_0 \omega_n \left(\zeta + \sqrt{\zeta^2 - 1}  \right) }{2 \omega_n \sqrt{\zeta^2 - 1}} e^{-\omega_n \left(\zeta - \sqrt{\zeta ^2 - 1}  \right)t}\\
X(t) &= \frac{e^{-\omega_n\zeta t}}{2\sqrt{\zeta^2 - 1}} \left[\left( - \frac{\dot{X}_0}{\omega_n} -X_0\left(\zeta - \sqrt{\zeta ^2 - 1}  \right)\right)e^{-\omega_n\sqrt{\zeta^2-1}t} + \left(\frac{\dot{X}_0}{\omega_n} + X_0\left(\zeta + \sqrt{\zeta ^2 - 1}  \right) \right)e^{\omega_n\sqrt{\zeta^2-1}t}\right]
\end{align*}



```python
#CASO 3 zeta>1 Sobreamortiguado
z=1.5
k=500
m=100
x0=3
v0=9
t=np.linspace(0,20,1000)
wo=np.sqrt(k/m)

x3=(np.exp(-z*w0*t)/(2*np.sqrt(z**2-1)))*((((v0/w0)+x0*(z+np.sqrt(z**2-1)))*np.exp(w0*t*np.sqrt(z**2-1)))+((-(v0/w0)+x0*(-z+np.sqrt(z**2-1)))*np.exp(-w0*np.sqrt(z**2-1)*t)))

plt.plot(t,x3,color='k')
plt.xlim((0, 10))
plt.xlabel('tiempo (s)')  
plt.ylabel('Amplitud')
plt.title('Sístema sobreamortiguado')
```
